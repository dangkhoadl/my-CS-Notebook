#### range(a,b) in C++

```C++
// Example: range(10,15) = [10, 11, 12, 13, 14]. Usage:
    // for(int &i:range<int>(10,15))
    //      cout << i << endl;
template <typename T>
vector<T> range(T N1, T N2) {
    vector<T> numbers(N2-N1);
    iota(numbers.begin(), numbers.end(), N1);
    return numbers;
}
```

#### slice(m,n) in C++

```cpp
// slice in range: [m, n)
template<typename T>
vector<T> slice_(const vector<T> &v, int m, int n) {
    auto first = v.cbegin() + m;
    auto last = v.cbegin() + n;
    return vector<T>(first, last);
}


// Usage
vector<int> v = {0, 1, 2, 3, 4, 5};
vector<int> res = slice_<int>(v, 2, 4);
    // [2, 3]
```

#### Erase

```C++
vector<char> A = {'0', '1', '2', '3', '4', '5'};
int start = 0, end = 3;
int pos = 2;

// Erase A[it->pos]
A.erase(A.begin() + pos);

// Erase A[it->start, it->end)
A.erase(A.begin() + start, A.begin() + end);
```
