# 3. Monotonic stack
- the stack is **decreasing** (or **increasing**) from bottom to top.


```Cpp
// Mono-decreasing Stack

// Before: [5,4,2,1]
// To push 3, we need to pop smaller (or equal) elements first
// After: [5,4,3]

vector<int> stack;
for(int num: nums) {
    // Pop stack
    while(!stack.empty() && stack.back() < num) stack.pop();
    
    // Push stack
    stack.push_back(num);
}
```
