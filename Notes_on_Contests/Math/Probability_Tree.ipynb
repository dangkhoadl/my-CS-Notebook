# [Archer](https://codeforces.com/gym/306143/problem/B)
- Give 4 Number: `A,B` and `C,D`
    + Archer SmallR has the probability of successful shooting the target each time: $\frac{A}{B}$
    + Archer Zanoes has the probability of successful shooting the target each time: $\frac{C}{D}$
- Find the probability that SmallR wins the competition
    + Each archer try to shoot the target in turns
    + SmallR go 1st
    + The one who shoots in the target first = winner

#### Constraints
- absolute error not exceed $10^{-6}$ 

#### Example

```
// Input
1 2 1 2

// Output
0.666666666667

```

# Solution
#### Simulation O(1e6)
- Let `p` = $\frac{A}{B}$
- Let `q` = $\frac{C}{D}$
- Name `X` = the probability SmallR win in all possible situation
    + $X = p + (1-p)(1-q)p + (1-p)^2(1-q)^2p + (1-p)^3(1-q)^3p + \dots$

- Code

```C++
int A,B,C,D;

double solve() {
    double p = (double)A/B;
    double q = (double)C/D;
 
    double ans = p;
    double tmp = (1.0-p) * (1.0-q);
    pr(tmp)
    for(int i=1; i<100000; ++i) {
        ans += tmp * p;
        tmp *= (1.0-p) * (1.0-q);
    }
 
    return ans;
}
```

#### Math simplified
- Name `X` = the probability SmallR win in all possible situation. The tree become


$$\begin{split}
    X &= p + (1-p)(1-q)X \\
 => X &= \frac{p}{1 - (1-p)(1-q)}
\end{split}$$

- Code

```Cpp
int A,B,C,D;

 
double solve() {
    double p = (double)A/B;
    double q = (double)C/D;
 
    return p / (1.0 - (1.0-p)*(1.0-q));
}
```
