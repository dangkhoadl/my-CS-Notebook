## [Longest Palindromic Substring](https://leetcode.com/problems/longest-palindromic-substring/)
- Given a string s, find the **longest** palindromic substring in s.
- Examples

```
Input: "babad"
Output: "bab"

Input: "cbbd"
Output: "bb"
```


#### Solution - $O(n^2)$
- If s[i,j] is palindrome -> s[i+1,j-1] palindrome
- Base cases:
    + s[i]: palindrome length = 1
    + s[i]s[j]: palindrome length = 2 if s[i]==s[j]

```C++
class Solution {
public:
    int N;
    vector<vector<int>> dp;
    int max_=1,i_=0,j_=0;
    int get(const string &s, int i, int j) {
        // Base cases
        if(i<0 || i>=N || j<0 || j>=N || i>j) return 0;
        if(i == j) return 1;                        // Case: palindrome a
        if(dp[i][j] != -1) return dp[i][j];

        // Search with memorization - O(n^2)
        get(s, i+1, j);
        get(s, i, j-1);
        int sub_pal = get(s, i+1, j-1);

        // Case: palindrome aa
        if(j==i+1 && s[i]==s[j]) 
            dp[i][j] = 2;

        // Case: palindrome bxb
        else if(s[i] == s[j] && sub_pal>0) 
            dp[i][j] = 2 + sub_str;

        // Case: not palindrome
        else
            dp[i][j] = 0;

        // Relaxation
        if(max_ < dp[i][j]) {
            max_ = dp[i][j];
            i_ = i;
            j_ = j;
        }
        return dp[i][j];
    }

    string longestPalindrome(const string &s) {
        N = s.size();
        dp.assign(N, vector<int>(N,-1));

        get(s,0,N-1);

        string res;
        for(int c=i_; c<=j_; ++c) res.push_back(s[c]);
        return res;
    }
};
```

#### Solution O($n^2$) - BottomUp
- Check s[j,i] palindrome O(n^3)
- Optimize to O(n^2):
    + Dont need to scan all j in range `[0,i]`
    + Choose `j = i-current_max_len-1` and `j = i-current_max_len` 

```C++
class Solution {
public:
    string s;
    bool check_palindrome(int l, int r) {
        while(l<r) {
            if(s[l] != s[r]) return false;
            l += 1;
            r -= 1;
        }
        return true;
    }
    string longestPalindrome(const string &str) {
        s = str;
        int N = s.size();
        if(N == 0) return "";

        int max_len = 1;
        int start_pos = 0;

        // Consider string s[j,i]
        for(int i=0; i<N; ++i) {
            // Case: bxb
            int j = i-max_len-1;
            if(j >= 0 && check_palindrome(j, i) ) {
                start_pos = j;
                max_len += 2;
                continue;
            }

            // Case: aa
            j = i - max_len;
            if(j >= 0 && check_palindrome(j, i) ) {
                start_pos = j;
                max_len += 1;
            }
        }
        return s.substr(start_pos, max_len);
    }
};
```
