# [Continuous Subarray Sum](https://leetcode.com/problems/continuous-subarray-sum/)

- **Given**
    + Array: `A`[0:N-1]
    + int: `k`
- **Return**: `true` if
    + Existing a subarray size >= 2
    + subarray sum % k = 0
    
#### Examples
```
Input: nums = [23,2,4,6,7], k = 6
Output: true
Explanation: [2, 4] is a continuous subarray of size 2 whose elements sum up to 6.

Input: nums = [23,2,6,4,7], k = 6
Output: true
Explanation: [23, 2, 6, 4, 7] is an continuous subarray of size 5 whose elements sum up to 42 (42 % 6 == 0)

Input: nums = [23,2,6,4,7], k = 13
Output: false
```

## Solution
- **Idea**
    + Subarray -> prefixsum
    + We have
$$
\begin{split}
    &(\text{prefix_sum}[r] - \text{prefix_sum}[l]) \% k = 0 \\
    => &\text{prefix_sum}[r] \% k = \text{prefix_sum}[l] \% k
\end{split}
$$

#### Code

```C++
class Solution {
public:
    bool checkSubarraySum(vector<int> &A, int k) {
        if(A.size() < 2) return false;

        // Build prefix sum
        vector<int> prefix_sum(A.size(), 0);
        for(int i=0; i<A.size(); ++i) {
            if(i == 0) prefix_sum[i] = A[i] % k;
            else prefix_sum[i] = (prefix_sum[i-1] + A[i]) % k; 
        }

        // [value]: l
        unordered_map<int, int> previous;
        previous[0] = -1;
        for(int r=0; r<A.size(); ++r) {
            // Check true condition
            if(previous.count(prefix_sum[r]) != 0 && r - previous[prefix_sum[r]] > 1) return true;

            // Greedy update: prefer smallest l
            if(previous.count(prefix_sum[r]) == 0) previous[prefix_sum[r]] = r;
        }

        return false;
    }
};
```

#### Simplified

```C++
class Solution {
public:
    bool checkSubarraySum(vector<int> &A, int k) {
        if(A.size() < 2) return false;

        // [value]: l
        unordered_map<int, int> previous;
        previous[0] = -1;

        int runner = 0;
        for(int r=0; r<A.size(); ++r) {
            // Update runner
            runner = (runner + A[r]) % k;

            // Check true condition
            if(previous.count(runner) != 0 && r - previous[runner] > 1) return true;

            // Greedy update: prefer smallest l
            if(previous.count(runner) == 0) previous[runner] = r;
        }

        return false;
    }
};
```
