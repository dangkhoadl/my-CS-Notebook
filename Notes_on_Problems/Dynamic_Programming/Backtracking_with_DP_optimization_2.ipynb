# Problem
- Given a stock chart: A[i], i in range [0, N-1]
- Within K transaction, determine the maximum profit
    + 1 transaction = 1 buy, 1 sell
    + 2 tranctions cant be paralelled, overlapped

#### Input:
```
K N
A[0] A[1] ... A[N-1]
```

#### Example
- Test set 1

```
Input:
2 6
10 22 5 75 65 80

Output:
87

Explanation:
[10 22] [5 75 65 80]
```

- Test set 2

```
Input:
3 4
20 580 420 900

Output:
1040

Explanation:
[20 580] [420 900]
```

- Test set 3

```
Input:
1 5
100 90 80 50 25

Output:
0

Explanation:
stock keep going down, no buy-sells
```

# Solution backtrack

```cpp

int K, N;
vector<int> A;


// 0: sold, 1: bought
int ans, res;
void get(int i, int pre, int k) {
    if(k <= K) {
        ans = max(ans, res);
    }
    if(i == N) return;

    // sold
    if(pre == 0) {
        // buy
        res -= A[i];
        get(i+1, 1, k);
        res += A[i];

        // keep
        get(i+1, 0, k);

        return;
    }

    // bought
    if(pre == 1) {
        // sell
        res += A[i];
        get(i+1, 0, k+1);
        res -= A[i];

        // keep
        get(i+1, 1, k);
        return;
    }
}

void sol() {
    ans = 0;
    res = 0;
    get(0, 0, 0);

    cout << ans << endl;
}
```

# Solution Backtrack with DP optimization

```cpp
int K, N;
vector<int> A;


// 0: sold, 1: bought
int ans;
int dp[33][3][13];
void get(int i, int pre, int k) {
    if(k <= K) {
        ans = max(ans, dp[i][pre][k]);
    }
    if(i == N) return;

    // sold
    if(pre == 0) {
        // buy
        if(dp[i+1][1][k] < dp[i][pre][k] - A[i]) {
            dp[i+1][1][k] = dp[i][pre][k] - A[i];
            get(i+1, 1, k);
        }

        // keep
        if(dp[i+1][0][k] < dp[i][pre][k]) {
            dp[i+1][0][k] = dp[i][pre][k];
            get(i+1, 0, k);
        }

        return;
    }

    // bought
    if(pre == 1) {
        // sell
        if(dp[i+1][0][k+1] < dp[i][pre][k] + A[i]) {
            dp[i+1][0][k+1] = dp[i][pre][k] + A[i];
            get(i+1, 0, k+1);
        }

        // keep
        if(dp[i+1][1][k] < dp[i][pre][k]) {
            dp[i+1][1][k] = dp[i][pre][k];
            get(i+1, 1, k);
        }
        return;
    }
}

void sol() {
    ans = 0;
    for(int i=0; i<=N; ++i)
        for(int d=0; d<=1; ++d)
            for(int k=0; k<=K; ++k) {
                dp[i][d][k] = INT_MIN;
    }

    dp[0][0][0] = 0;
    get(0, 0, 0);

    cout << ans << endl;
}
```
