# Problem
- Given A list of connected dominos as `domino_front -> domino_back`
    + If front is knocked -> back will also falls

- Calculate the minimum number of knock by hand to push down all dominos

- Example
```
0 -> 1
1 -> 2
  // Knock 1 time: 0 -> 1 -> 2
```


# Solution
- Build a directed graph from the list of dominos
- dfs 2 times
    + 1: construct topo-order
    + 2: count the number of knock by following topo-order

#### Code

```cpp
int V, E;
vector<int> edges[100003];

// dfs1
bool vis[100003];
vector<int> topoOrder;
void explore(int v) {
    vis[v] = true;

    for(int &w:edges[v]) {
        if (!vis[w]) {
            explore(w);
        }
    }

    topoOrder.push_back(v);
}
void get_topo() {
    memset(vis, false, sizeof(vis));

    // Get topo order
    topoOrder.clear();
    for (int v=0; v<V; ++v) {
        if(!vis[v]) {
            explore(v);
        }
    }

    reverse(topoOrder.begin(), topoOrder.end());
}

// dfs2
void get(int v) {
    vis[v] = true;
    for(int &w:edges[v]) {
        if (!vis[w]) {
            get(w);
        }
    }
}
int dfs() {
    memset(vis, false, sizeof(vis));
    int ans = 0;
    for(int &v:topoOrder) {
        if(!vis[v]) {
            ans += 1;
            get(v);
        }
    }
    return ans;
}
int sol() {
    // dfs 1
    get_topo();

    // dfs 2
    return dfs();
}
```
