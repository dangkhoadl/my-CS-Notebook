# Assign Task - 1 person
#### Problem
- Have a list of tasks
    + Each task has [start, end] time
- Have 1 worker: Do the **maximum number of tasks** as possible

#### Solution

```C++
struct Task {
    int start_;
    int end_;
};
int N;
vector<Task> tasks;


int sol() {
    // Sort tasks by end time increasing
    sort(tasks.begin(), tasks.end(),
        [](const Task &a, const Task &b) {
            return a.end_ < b.end_;
        });

    // Assign tasks
    int cnt = 0;
    int worker = 0;
    for(auto &task: tasks) {
        if(worker <= task.start_) {
            worker = task.end_;
            cnt += 1;
        }
    }
    return cnt;
}
```

# [Assign Task: 2 persons](https://codingcompetitions.withgoogle.com/codejam/round/000000000019fd27/000000000020bdf9)
#### Problem
- Have a list of tasks
    + Each task has [start, end] time
- Have 2 persons: Camie and Jamie
- Assign all the tasks to Camie and Jamie
    + No overlapped tasks
    + **All tasks must be assigned** to either Camie or Jamie

#### Solution

```C++
struct Task {
    int id_;
    int start_;
    int end_;
};
int N;
vector<Task> tasks;


string sol() {
    // Sort tasks by start time increasing
    sort(tasks.begin(), tasks.end(),
        [](const Task &a, const Task &b) {
            return a.start_ < b.start_;
        });

    // Assign tasks
    string res(N, '-');
    int Camie = 0, Jamie = 0;
    for(auto &task: tasks) {
        // Assign task to Camie if Camie free
        if (Camie <= task.start_) {
            Camie = task.end_;
            res[task.id_] = 'C';
        }
        // Assign task to Jamie if Camie not free, Jamie free
        else if(Jamie <= task.start_) {
            Jamie = task.end_;
            res[task.id_] = 'J';
        }
        // Impossible to assign tasks
        else {
            return "";
        }
    }
    return res;
}
```
