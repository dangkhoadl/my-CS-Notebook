# Cab Driver Problem - [ Unique Paths](https://leetcode.com/problems/unique-paths/)
- Go from A to B
    + m: horizontal blocks
    + n: vertical blocks
- Can only move Up or Right
- How many routes that a cab driver go from A->B?


<img src="./img/1.jpg" alt="drawing" width="300"/>

#### Dynamic Programming
- Path(A,B) = Path(A, $B_1$) + Path(A, $B_2$)
    - Base case
        + dp(m,0) = 1
        + dp(0,n) = 1
    - DP relation: `dp(m,n) = dp(m-1,n) + dp(m, n-1)`

<img src="./img/2.jpg" alt="drawing" width="300"/>


```C++
class Solution {
public:
    int uniquePaths(int N, int M) {
        vector<vector<int>> dp(N, vector<int>(M, 0));
        dp[0][0] = 1

        for(int i=0; i<N; ++i) {
            for(int j=0; j<M; ++j) {
                if(i == 0 && j == 0) continue;
                dp[i][j] = (i-1 >= 0 ? dp[i-1][j]: 0) + (j-1 >= 0 ? dp[i][j-1]: 0);
            }
        }
        return dp[N-1][M-1];
    }
};
```

#### Combinatorics
- the route from A to B = A set of Up and Right. Ex: **RUURURU**

<img src="./img/1.jpg" alt="drawing" width="300"/>


- Config
    - Path length = m + n
    - move R: m times
    - move U: n times
- Distribute `m` Rs into `(m+n)` containers **or** `n` Us into `(m+n)` containers


$$
\begin{split}
    \text{Path(A,B)} &= C_{m+n}^n = C_{m+n}^m \\
      &= \frac{(m+n)(m+n-1)\dots(m+1)}{n(n-1)\dots1}
\end{split}
$$

```C++
class Solution {
public:
    long long uniquePaths(int m, int n) {
        long long n_ = min(m,n) - 1;
        long long m_ = max(m,n) - 1;

        // Compute (m_ + n_) C n_
        long long ans = 1;
        for(long long i=1; i<=n_; ++i) {
            ans *= (m_ + i);
            ans /= i; 
        }
        return ans;
    }
};
```
