# [Frog Jump](https://www.facebook.com/reel/1137841151320003)
- A frog on any leaf can jump to other leaves with equal probability
    - at `B`: It can jump to A,C,D with P=1/3
- It starts at `A`
- Find the probability that after 6 jumps it back to `A`

<img src="img/frog_1.png" width="500" height="500">

# Solution
#### Modeling
- If the frog `at A`, it definitely jump to `not at A` with P = 1
- If the frog `not at A`
    + It will jump to `not at A` with P = 2/3
    + It will jump to `at A` with P = 1/3

<img src="img/frog_2.png" width="800" height="400">

#### Solve
- Let $A_i$ = the event the frog will jump to A after i-th jump
- We have
$$P(\overline{A_i}) = 1 - P(A_i)$$

- After the first jump, the frog will definitely be `not at A`
    + $P(A_1) = 0$ and $P(\overline{A_1}) = 1$
- For the frog to land at `A` for $A_n$, it must be `not at A` at $A_{n-1}$
$$\begin{split}
P(A_n) &= \frac{1}{3}P(\overline{A_{n-1}}) \\
    &= \frac{1}{3} \left[ 1 - P(A_{n-1})\right]
\end{split}$$


```python
def solve_frogjump(n):
    if n == 1: return 0
    return 1/3*(1 - solve_frogjump(n-1))

solve_frogjump(6)
```
