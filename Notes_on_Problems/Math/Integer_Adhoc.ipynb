# Problem
- N windows which provide the tickets to get entry into a park
    + There are A[i] people already queue at each window
    + Each window gives ticket to one person in one minute.
- Initially, John stands in front of the first window
    + After each minute, if he didnâ€™t get the ticket, he moves on to the next window
    + If he is at window 1, he will move to 2
    + If at 2nd, he will move to 3rd
    + If he is at last window, he will move to 1st
- Find the window number at which he will get the ticket.

#### Example

```
4
2 3 2 0
```

- Output: 2

- Explanation

```
John at Window 0: [2, 3, 2, 0]
John at Window 1: [1, 2, 1, 0]
John at Window 2: [0, 1, 0, 0]
So, when John is at window 2, he got zero people before him. Hence, he will get the ticket at window 2
```

## Solution
- Find the time that John will visit the i-th window
- solution = min (the visit time that this window running out of people)
    + $f(i) = i + k\ N$
        + $i + k\ N \geq A[i] \to k \leq \frac{A[i]-i}{N}$

```cpp
int N;
vector<int> A;

int ceil_(int m, int n) {
    assert(n != 0);

    int a = m/n+1;
    while(m <= n*a) --a;
    ++a;

    int b = (m+n)/n-1;
    while(n*b < (m+n)) ++b;
    --b;

    if(a==b)
        return a;
    assert(1==0);
}

int f(int i) {
    int k = ceil_(A[i]-i, N);
    return i+k*N;
}


void sol() {
    int ans = 0;
    int min_ = INT_MAX;
    for(int i=0; i<N; ++i) {
        int f_i = f(i);
        if(min_ > f_i) {
            min_ = f_i;
            ans = i;
        }
    }
    cout << ans << endl;
}
```
