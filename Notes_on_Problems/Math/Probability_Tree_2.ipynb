# [21 Game](https://leetcode.com/problems/new-21-game/)
- K: starts with 0 points, and draws a number while < K points.
- W: number range in `[1,W]`
- N: Calc probability that the final points <= N

#### Constraints
- $0 \leq K \leq N \leq 10^4$

#### Examples

```
Input: N = 10, K = 1, W = 10
Output: 1.00000

Input: N = 6, K = 1, W = 10
Output: 0.60000

Input: N = 21, K = 17, W = 10
Output: 0.73278
```

#### Solution O((K+W)* W)

```C++
class Solution {
public:
    int N, K, W;
    vector<double> dp;
    double get(int points) {
        // Base cases
        if(points >= K) {
            if(points <= N) return 1.0;
            return 0.0;
        }
        if(dp[points] >= 0.0) return dp[points];

        // Get
        double prob = 0.0;
        for(int i=1; i<=W; ++i) {
            prob += 1.0/W * get(points + i);
        }

        return dp[points] = prob;
    }
    double new21Game(int n, int k, int maxPts) {
        N = n, K = k, W = maxPts;
        dp.assign(K+W+1, -1.0);
        return get(0);
    }
};
```

#### Solution O(K+W)
- We have

$$\begin{cases}
f(x) &= \frac{1}{W}f(x+1) + \frac{1}{W}f(x+2) + \dots + \frac{1}{W}f(x+W) \\
f(x+1) &= \frac{1}{W}f(x+2) + \frac{1}{W}f(x+3) + \dots + \frac{1}{W}f(x+W) + \frac{1}{W}f(x+W+1)
\end{cases}$$

$$\begin{split}
=> f(x) - f(x+1) &= \frac{1}{W}f(x+1) - \frac{1}{W}f(x+W+1) \\
f(x) &= \frac{W+1}{W}f(x+1) - \frac{1}{W}f(x+W+1)
\end{split}$$

```Cpp
class Solution {
public:
    int N, K, W;
    vector<double> dp;
    double get(int points) {
        // Base cases
        if(points >= K) {
            if(points <= N) return 1.0;
            return 0.0;
        }
        if(dp[points] >= 0.0) return dp[points];

        // Exception case: when points + 1 >= K, we compute normally
        if(points + 1 >= K) {
            double prob = 0.0;
            for(int i=1; i<=W; ++i) {
                prob += (points + i <= N) ? 1.0/W : 0;  
            }
            return dp[points] = prob;
        }
        
        // Math formula
        double prob = (W+1.0)/W*get(points + 1) - 1.0/W*get(points+1+W);

        return dp[points] = prob;
    }
    double new21Game(int n, int k, int maxPts) {
        N = n, K = k, W = maxPts;
        dp.assign(K+W+1, -1.0);
        return get(0);
    }
};
```
