# [house-robber-iii](https://leetcode.com/problems/house-robber-iii/)
- **Given**: A Binary Tree
- **Find** the maximum sum values of a subset of nodes such that
    + No nodes in the subset adjacent (not child-parent relation)
    
#### Example 1

<img src="./img/2.jpg" width="200"/>

```
Output: 7
Explanation: Maximum = 3 + 3 + 1 = 7.
```

#### Example 2

<img src="./img/3.jpg" width="200"/>

```
Output: 9
Explanation: Maximum = 4 + 5 = 9.
```

#### Example 3

<img src="./img/4.jpg" width="150"/>

```
Output: 7
Explanation: Maximum = 4 + 3 = 7.
```

## Solution
#### idea
- Same as select/skip DP, each node can be skipped or added

#### Code

```Cpp
class Solution {
public:
    pair<int, int> get(TreeNode* cur) {
        if(cur == NULL) return {0, 0};

        auto [skip_left, add_left] = get(cur->left);
        auto [skip_right, add_right] = get(cur->right);

        // Scenarios 1: skip cur, left/right child can be skiped or added, select best case
        int skip_cur = max({
            add_left + add_right,
            add_left + skip_right,
            skip_left + add_right,
            skip_left + skip_right});

        // Scenarios 2: add cur, must skip both left,right child
        int add_cur = cur->val + skip_left + skip_right;

        return {skip_cur, add_cur};
    }
    int rob(TreeNode* root) {
        auto [skip_root, add_root] = get(root);
        return max(skip_root, add_root); 
    }
};
```

# [Maximum Path Sum](https://leetcode.com/problems/binary-tree-maximum-path-sum/)

- **Given**: A Binary Tree
- **Find** the maximum path sum
    + Each pair of adjacent nodes in the sequence has an edge connecting them 
    + Each node only appear in the sequence at most once
    
#### Example 1

<img src="./img/5.jpg" width="300"/>

```
Output: 6
Explanation: The optimal path is 2 -> 1 -> 3 with a path sum of 2 + 1 + 3 = 6
```

#### Example 2

<img src="./img/6.jpg" width="300"/>

```
Output: 42
Explanation: The optimal path is 15 -> 20 -> 7 with a path sum of 15 + 20 + 7 = 42.
```

## Solution
#### Code

```C++
class Solution {
public:
    const int ninf = -9999;
    pair<int, int> get(TreeNode* cur) {
        if(cur == NULL) return {ninf, ninf};

        auto [left_path_complete, left_path_not_complete] = get(cur->left);
        auto [right_path_complete, right_path_not_complete] = get(cur->right);

        // Path completed at cur
        int cur_path_complete = max({
            left_path_complete, left_path_not_complete,
            right_path_complete, right_path_not_complete,
            cur->val + left_path_not_complete + right_path_not_complete
        });

        // Path not completed at cur
        int cur_path_not_complete = max({
            cur->val + left_path_not_complete,
            cur->val + right_path_not_complete,
            cur->val
        });

        return {cur_path_complete, cur_path_not_complete};
    }
    int maxPathSum(TreeNode* root) {
        auto [root_path_complete, root_path_not_complete] = get(root);
        return max(root_path_complete, root_path_not_complete);
    }
};
```
