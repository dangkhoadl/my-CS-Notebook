# [K-th Symbol in Grammar](https://leetcode.com/problems/k-th-symbol-in-grammar/)
- **Given**
    + `N`: Row (base 1)
    + `K`: index (base 1)
- Generate a string follow thise rule:
    + Start with row 1 = "0"
    + For row i-th
        + with each `0` replace by `01`
        + with each `1` replace by `10`

```
row 1: 0 
row 2: 01 
row 3: 0110 
row 4: 01101001
```

- **Return**: the value at `K-th` index (base 1) at row `N-th`

#### Constraints
- $1 \leq N \leq 30$
- $1 \leq K \leq 2^{N-1}$

#### Examples

```
Input: n = 1, k = 1
Output: 0
Explanation:
    row 1: [0]
```

```
Input: n = 2, k = 1
Output: 0
Explanation: 
    row 1: 0
    row 2: [0]1
```

```
Input: n = 2, k = 2
Output: 1
Explanation: 
    row 1: 0
    row 2: 0[1]
```

## Solution
- **Insight**: the whole structure is a **full binary tree**

```
row 1:            0
             /         \
row 2:      0           1
          /   \       /   \
row 3:   0     1     1      0
        / \   / \   / \    / \
row 4: 0   1 1   0 1   0  0   1
```

- Convert k to tree idx then solve
<img src="./img/FullBinary_2.jpg" width="300"/>

$$\begin{split}
\text{tree_idx} &= (\text{array_size}-1) + \text{array_idx} \\
&= (2^{h-1}-1) + \text{array_idx}
\end{split}$$

```Cpp
class Solution {
public:
    int get(int i) {
        if(i == 0) return 0;

        // get par
        int par = (i-1) / 2;
        int par_val = get(par);

        // par_val = 0, cur = left child (odd)
        if(par_val == 0 && i%2 == 1) return 0;

        // par_val = 0, cur = right child (even)
        else if(par_val == 0 && i%2 == 0) return 1;

        // par_val = 1, cur = left child (odd)
        else if(par_val == 1 && i%2 == 1) return 1;

        // par_val = 1, cur = right child (even)
        else if(par_val == 1 && i%2 == 0) return 0;

        assert(false);
        return -1;
    }

    int kthGrammar(int N, int K) {
        // base 1 -> base 0
        int k = K - 1;

        // array idx -> tree idx
        int i = (1<<N-1) - 1 + k;
        return get(i);
    }
};
```
