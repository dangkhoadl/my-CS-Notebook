## [Parsing A Boolean Expression](https://leetcode.com/problems/parsing-a-boolean-expression/)
- Return the result of a boolean expression
- An expression can either be:
    - "t", evaluating to True;
    - "f", evaluating to False;
    - "!(expr)", evaluating to the logical NOT of the inner expression expr;
    - "&(expr1,expr2,...)", evaluating to the logical AND of 2 or more inner expressions expr1, expr2, ...;
    - "|(expr1,expr2,...)", evaluating to the logical OR of 2 or more inner expressions expr1, expr2, ...
- **Examples**

```
Input: expression = "!(f)"
Output: true

Input: expression = "|(f,t)"
Output: true

Input: expression = "&(t,f)"
Output: false

Input: expression = "|(&(t,f,t),!(t))"
Output: false
```

#### Solution

```C++
class Solution {
private:
    string S;
    int N;
public:
    vector<bool> solve_expr_in_paren(int &i) {
        vector<bool> exprs;
        for(;i<N;++i) {
            // '&', '|', '!'
            char op;
            if(i<N && S[i] == '&' || S[i] == '|' || S[i] == '!') op = S[i++];

            if(i<N && S[i] == '(') {
                // eval
                vector<bool> paren_exprs = solve_expr_in_paren(++i);
                bool res;
                if(op == '&') {
                    res = true;
                    for(bool x:paren_exprs) res = res && x;
                }
                else if(op == '|') {
                    res = false;
                    for(bool x:paren_exprs) res = res || x;
                }
                else if(op == '!') {
                    // Assure !() has 1 operant
                    assert(paren_exprs.size() == 1);
                    res = !paren_exprs[0];
                }

                // Update
                exprs.push_back(res);
            }
            else if(i<N && S[i] == ')') {
                return exprs;
            }
            else if(i<N && S[i] == 't') {
                exprs.push_back(true);
            }
            else if(i<N && S[i] == 'f') {
                exprs.push_back(false);
            }
        }
        return exprs;
    }
    bool parseBoolExpr(const string &expression) {
        S = expression;
        N = S.size();
        int i = 0;

        vector<bool> ans = solve_expr_in_paren(i); 
        assert(ans.size() == 1);
        return ans[0];
    }
};
```
