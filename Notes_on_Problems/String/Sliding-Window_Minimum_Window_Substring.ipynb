## [Minimum Window Substring](https://leetcode.com/problems/minimum-window-substring/)
- Given a string s and a string t, find the minimum window in S which will contain all the characters in T in complexity O(n).
- Example

```
Input: S = "ADOBECODEBANC", T = "ABC"
Output: "BANC"

Input: S = "aa", T = "aa"
Output: "aa"
```

#### Solution
- Ideas
    + window: 128 ASCII character
        + character in t: [0:num]
        + character not in t: [-inf:0]

```C++
class Solution {
public:
    string minWindow(const string &s, const string &t) {
        // Preprocess 
        vector<int> window(128, 0);
        for(const auto &c: t) window[c] += 1;
        int miss = t.size();

        // Prreprocess ans
        int len = INT_MAX;
        int a_ = 0;

        // Sliding window
        int a = 0;
        int b = 0;
        int N = s.size();
        while(b < N) {
            // Increase(b) as usual
            if(window[s[b]] > 0) --miss;
            window[s[b]] -= 1;
            ++b;

            // Increase(a) if substring match
            while(miss == 0) {
                // Relax
                if(b-a < len)  {
                    a_ = a;
                    len = b-a;
                }

                // Increase(a)
                if(window[s[a]] == 0) ++miss;
                window[s[a]] += 1;
                ++a;
            } 
        }

        // Return result
        return len==INT_MAX ? "" : s.substr(a_, len);
    }
};
```
