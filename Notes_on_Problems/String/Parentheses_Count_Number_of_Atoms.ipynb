## [Number of Atoms](https://leetcode.com/problems/number-of-atoms/)

- **Given**
    + A chemical formula (`Mg(OH)2`)
    - An atomic element always starts with an uppercase character, then zero or more lowercase letters, representing the name. (H, Mg)
  
- **Return**: the count of each atom (`H2MgO2`)
    + Sorted by alphabetic order
    + The count value = empty if count = 1

#### Example
```
Input: formula = "H2O"
Output: "H2O"
Explanation: The count of elements are {'H': 2, 'O': 1}.


Input: 
formula = "Mg(OH)2"
Output: "H2MgO2"
Explanation: The count of elements are {'H': 2, 'Mg': 1, 'O': 2}.


Input: formula = "K4(ON(SO3)2)2"
Output: "K4N2O14S4"
Explanation: The count of elements are {'K': 4, 'N': 2, 'O': 14, 'S': 4}.


Input: formula = "(H)"
Output: "H"
```

#### Solution

```C++
class Solution {
private:
    string S;
    int N;
    bool isLower(char c) {
        return 'a' <= c && c <= 'z';
    }
    bool isUpper(char c) {
        return 'A' <= c && c <= 'Z';
    }
public:
    map<string, int> solve_paren(int &i) {
        map<string, int> res;
        while(i<N) {
            // Solve non (): H2O
            if(isUpper(S[i])) {
                //Get elem: Mg
                string elem = "";
                elem += S[i++]; // M
                while(i<N && isLower(S[i])) elem += S[i++]; // g

                // Get count: O2 or O
                string num = "";
                while(i<N && isdigit(S[i])) num += S[i++];
                int cnt = (num == "") ? 1 : stoi(num);

                // Update cnt
                res[elem] += cnt;
            }
            else if(S[i] == '(') {
                // Solve (...)
                map<string, int> paren_res = solve_paren(++i);
                i += 1; // Skip ')'

                // Case: (...)2 or (...)
                string num = "";
                while(i<N && isdigit(S[i])) num += S[i++];
                int multiplier = (num == "") ? 1 : stoi(num);
                
                // Update res from (paren_res)
                for(auto &[elem_, cnt_]: paren_res) {
                    res[elem_] += multiplier*cnt_;
                }
            }
            else if(S[i] == ')') {
                return res;
            }
        }
        return res;
    }
    string countOfAtoms(const string &formula) {
        S = formula;
        N = S.size();
        int i = 0;

        // Solve
        map<string, int> res = solve_paren(i);

        // Parse ans
        string ans = "";
        for(auto &[elem_, cnt_]: res) {
            ans += elem_;
            ans += (cnt_ == 1) ? "" : to_string(cnt_);
        }
        return ans;
    }
};
```
