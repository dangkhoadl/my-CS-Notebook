# [Problem](https://leetcode.com/problems/decode-string/)

- **Given**: encoded string
    + Format: `k[encoded_string]`
- **Return** decoded string.
    + k \* encoded_string 

#### Examples

```
Input: s = "3[a]2[bc]"
Output: "aaabcbc"
Explanation: (a)(a)(a)(bc)(bc)

Input: s = "3[a2[c]]"
Output: "accaccacc"
Explanation: 3[a2[c]] --> 3[a(c)(c)] --> (acc)(acc)(acc)

Input: s = "2[abc]3[cd]ef"
Output: "abcabccdcdcdef"
Explanation: (abc)(abc)(cd)(cd)(cd)ef
```

# Solution
#### Stack

```C++
class Solution {
private:
    bool ischar(char c) {
        return 'a' <= c && c <= 'z';
    }
public:
    string decodeString(const string &S) {
        int N = S.size();
        int i = 0;

        string ans = "";
        vector<string> my_stack;
        for(;i < N; ++i) {
            // Parse string
            //      Note: Since there is a case "3[a2[c]]" --> Prior parsing string
            while(i<N && ischar(S[i])) {
                ans += S[i++];
            }

            // Parse num
            string num = "";
            while(i<N && isdigit(S[i])) {
                num += S[i++];
            }

            if(i<N && S[i] == '[') {
                // Use stack as a time-based memory
                my_stack.push_back(ans);
                my_stack.push_back(num);

                ans = "";
                num = "";
            }
            else if(i<N && S[i] == ']') {
                // Pop stack
                int prev_num = stoi(my_stack.back());
                my_stack.pop_back();

                string pre_string = my_stack.back();
                my_stack.pop_back();

                // ans = pre_string + prev_num*ans;
                string tmp = "";
                while(prev_num--) tmp += ans; 
                ans = pre_string + tmp;
            }
        }
        return ans;
    }
};
```

#### Recursion
- **Strategy**
    + Treat `(...)` as a subproblem
    + Meet `'('`: `ans = ans + solve_paren(...)`
    + Meet `')'`: return `ans`

```C++
class Solution {
private:
    string S;
    int N;

    bool ischar(char c) {
        return 'a' <= c && c <= 'z';
    }
public:
    string solve_parenthesis(int &i) {
        string res = "";
        for(;i<N;++i) {
            // Parse string
            //      Note: Since there is a case "3[a2[c]]" --> Prior parsing string
            while(i<N && ischar(S[i])) {
                res += S[i++];
            }

            // Parse num
            string num = "";
            while(i<N && isdigit(S[i])) {
                num += S[i++];
            }


            if(i<N && S[i] == '[') {
                string string_in_paren = solve_parenthesis(++i);

                // res = res + num*string_in_paren;
                int cnt = stoi(num);
                string tmp = "";
                while(cnt--) tmp += string_in_paren;

                res += tmp;
            }
            else if(i<N && S[i] == ']') {
                return res;
            }
        }
        return res;
    }
    string decodeString(const string &Str) {
        S = Str;
        N = S.size();
        int i = 0;

        return solve_parenthesis(i);
    }
};
```
