# [Minimum Edit Distance](https://leetcode.com/problems/edit-distance/)
- Edit distance measures the similarity of two strings.
- The edit distance between two strings is the minimum number of operations:
    + Insert a character (cost = 1)
    + Delete a character (cost = 1)
    + Replace a character (cost = 1)
    
#### Examples

```
Input: s1 = "horse", s2 = "ros"
Output: 3
Explanation: 
    horse -> rorse (replace 'h' with 'r')
    rorse -> rose (remove 'r')
    rose -> ros (remove 'e')
    
Input: s1 = "intention", s2 = "execution"
Output: 5
Explanation: 
    intention -> inention (remove 't')
    inention -> enention (replace 'i' with 'e')
    enention -> exention (replace 'n' with 'x')
    exention -> exection (replace 'n' with 'c')
    exection -> execution (insert 'u')
```


## Solution

-  Bottom up DP $O(n^2)$

```C++
class Solution {
public:
    int minDistance(string &A, string &B) {
        int n = A.size();
        int m = B.size();

        // dp[i+1][j+1] marks the optimal sol of prefix A[:i], B[:j] 
        vector<vector<int>> dp(n+1, vector<int>(m+1, INT_MAX));
        dp[0][0] = 0;

        // Left to right
        for(int i=0; i<=n; ++i) for(int j=0; j<=m; ++j) {
            // Bottom up: propagate dp[i][j] -> dp[i+1][j],dp[i][j+1],dp[i+1][j+1]
            int cur = dp[i][j]; 
            if(cur == INT_MAX) continue;

            // Case similar: Not edit
            if(A[i] == B[j]) {
                if(i+1 <= n && j+1 <=m) dp[i+1][j+1] = min(dp[i+1][j+1], cur);
            }

            // Not similar
            else {
                // Case insert A
                if(i+1 <= n) dp[i+1][j] = min(dp[i+1][j], cur + 1);
                
                // Case Delete A (Insert B)
                if(j+1 <= m) dp[i][j+1] = min(dp[i][j+1], cur + 1);

                // Case Substitute
                if(i+1 <= n && j+1 <= m) dp[i+1][j+1] = min(dp[i+1][j+1], cur + 1);
            }
        }

        return dp[n][m];
    }
};
```
