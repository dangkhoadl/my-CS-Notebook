## [Brace Expansion II](https://leetcode.com/problems/brace-expansion-ii/)
- Parse the expression with following rules
    - Rule 1: For every lowercase letter x, we have R(x) = {x}
        + R("{a,b,c}") = {"a","b","c"}
    - Rule 2: For expressions $e_1, e_2, ... , e_k$ with k >= 2, we have $R({e_1,e_2,...}) = R(e_1) ∪ R(e_2) ∪ ...$
        + R("{{a,b},{b,c}}") = {"a","b","c"}
    - Rule 3: For expressions $e_1\ and\ e_2$, we have $R(e_1e_2)\ =\ {a + b,\ for\ (a, b)\ in\ R(e_1)\ ×\ R(e_2)}$
        + \+ denotes concatenation
        + × denotes the cartesian product
        + R("{a,b}{c,d}") = {"ac","ad","bc","bd"}
        
- **Examples**

```
Input: "{a,b}{c,{d,e}}"
Output: ["ac","ad","ae","bc","bd","be"]

Input: "{{a,z},a{b,c},{ab,z}}"
Output: ["a","ab","ac","z"]
```

#### Solution

```C++
class Solution {
public:
    int N;
    string S;

    // {{a,b},{b,c}} -> {a,b,c}
    set<string> get_2(int &i) {
        set<string> res;
        res = get_3(i);
        while(i < N) {
            if(S[i] != ',') break;

            ++i;
            set<string> tmp = get_3(i);
            res.insert(tmp.begin(), tmp.end());
        }
        return res;
    }

    // {a,b}{c,d} --> {"ac","ad","bc","bd"}
    // ad{b,c} --> {"adb", "adc"}
    set<string> merge(const set<string> &x1, const set<string> &x2) {
        if(x1.size() == 0) return x2;
        if(x2.size() == 0) return x1;

        set<string> res;
        for(auto &e:x1) for(auto &f:x2) res.insert(e+f);
        return res;
    }
    set<string> get_3(int &i) {
        set<string> res;
        while(i < N) {
            if(S[i] == ',' || S[i] == '}') break;
            if(S[i] == '{') {
                ++i;
                set<string> tmp = get_2(i);
                res = merge(res, tmp);
                ++i;
            } else {
                string word = "";
                while(i < N && 'a' <= S[i] && S[i] <= 'z') {
                    word.push_back(S[i]);
                    ++i;
                }

                set<string> tmp;
                tmp.insert(word);
                res = merge(res, tmp);
            }
        }
        return res;
    }
    
    vector<string> braceExpansionII(string &exp) {
        this->S = exp;
        this->N = S.size();

        int i = 0;
        set<string> ans = get_2(i);
        return vector<string>(ans.begin(), ans.end());
    }
};
```
