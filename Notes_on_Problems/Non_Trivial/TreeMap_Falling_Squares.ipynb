# [Falling Squares](https://leetcode.com/problems/falling-squares/)

- Give a list of `N` squares
    + each square represent by `[left, side_length]`

- Each time we drop 1 square (0 -> N-1)
- Return the maximum global height after each drop

#### Example 1

```
Input: [[1, 2], [2, 3], [6, 1]]
Output: [2, 5, 5]
Explanation:
```

# Solution
- Maintain a list of `(x, max_height_at_x)`
- Update TreeMap everytime a cell drop (cur_l, cur_r)
- Init: `(0,0), (INT_MAX, 0)`

```C++
class Solution {
public:
    vector<int> fallingSquares(vector<vector<int>>& squares) {
        // [x] = h
        map<int,int> pre_heights = {{0,0}, {INT_MAX,0}};
        vector<int> ans;
        int max_height = 0;
        for(auto &cur_sq : squares){
            int cur_l = cur_sq[0];
            int cur_r = cur_sq[0] + cur_sq[1];
            int cur_h = cur_sq[1];

            // low, up = [l, r+], if l not exist: low, up = [l-, r+]
            auto low = pre_heights.lower_bound(cur_l);
            if(low->first != cur_l) low = prev(low);
            auto up = pre_heights.lower_bound(cur_r);

            // Find local local_pre_height in [l, r+] or [l-, r+], Update global ans
            int local_pre_height = 0;
            for(auto it=low; it!=up; ++it) {
                local_pre_height = max(local_pre_height, it->second);
            }
            max_height = max(max_height, local_pre_height + cur_h);
            ans.push_back(max_height);

            // tmp_h for r
            int tmp_h = (up->first == cur_r) ? up->second : prev(up)->second;

            // Erase [l+, r+]
            auto nex_low = next(low);
            pre_heights.erase(nex_low, up);

            // Update l, r
            pre_heights[cur_l] = local_pre_height + cur_h;
            pre_heights[cur_r] = tmp_h;
        }
        return ans;
    }
};
```
