## [Gas Station](https://leetcode.com/problems/gas-station/)
- There are N gas stations along a circular route
    - gas[i]: amount of gas at station i
    + cost[i]: amount of gas consumed to travel from station i to station i+1
- Return the starting gas station's index if you can travel around the circuit once in the clockwise direction, otherwise return -1.
- **Examples 1**

<img src="./img/2.png" alt="drawing" width="300"/>


```
Input: 
gas  = [1,2,3,4,5]
cost = [3,4,5,1,2]

Output: 3

Explanation:
    Start at station 3 (index 3) and fill up with 4 unit of gas. Your tank = 0 + 4 = 4
    Travel to station 4. Your tank = 4 - 1 + 5 = 8
    Travel to station 0. Your tank = 8 - 2 + 1 = 7
    Travel to station 1. Your tank = 7 - 3 + 2 = 6
    Travel to station 2. Your tank = 6 - 4 + 3 = 5
    Travel to station 3. The cost is 5. Your gas is just enough to travel back to station 3.
    Therefore, return 3 as the starting index.
```

- **Examples 2**

```
Input: 
gas  = [2,3,4]
cost = [3,4,3]

Output: -1

Explanation:
    You can't start at station 0 or 1, as there is not enough gas to travel to the next station.
    Let's start at station 2 and fill up with 4 unit of gas. Your tank = 0 + 4 = 4
    Travel to station 0. Your tank = 4 - 3 + 2 = 3
    Travel to station 1. Your tank = 3 - 3 + 3 = 3
    You cannot travel back to station 2, as it requires 4 unit of gas but you only have 3.
    Therefore, you can't travel around the circuit once no matter where you start.
```

#### Solution
- Observations
    + if $\sum\limits_{i=0}^{N-1} gas[i] < \sum\limits_{i=0}^{N-1} cost[i]$: Impossible to have a solution  
    => Return -1 if $total = \sum\limits_{i=0}^{N-1} gas[i] - cost[i] < 0$
    + A station i cant be the starting station if 
        + gas[i] - cost[i] + gas[i+1] - cost[i+1] + ... < 0 
- Solution
    + if total >= 0: There exists a solution
    + Remove all impossible solutions, the only solution left is the answer
        + Choose a starting station, if not feasible => choose the next station
        + Continue
    
```C++
class Solution {
public:
    int canCompleteCircuit(vector<int>& gas, vector<int>& cost) {
        int N = gas.size();
        int total = 0;
        int curr = 0;

        int N_s = 0;
        for(int i=0; i<N; ++i) {
            total += (gas[i]-cost[i]);
            curr += (gas[i]-cost[i]);

            if(curr < 0) {
                N_s = (i+1)%N;
                curr = 0;
            }
        }

        return total < 0 ? -1 : N_s;
    }
};
```

## Proof
#### if $total \geq	 0$ there always exist a solution
- Assume the existing a starting station $N_s$
- The algorithm above proves that it is possible to travel from $N_s$ to 0  
$gas[N_s] - cost[N_s] + gas[N_s+1] - cost[N_s+1] + ... \geq 0$
- Must prove that  it's always possible to travel from 0 to $N_s$. 
    + **Proof by contradiction**: 
        + Assume a station $k \neq N_s, 0 \leqslant k \leqslant N_s$, couldn't reach if starting from $N_s$
        + Prove that this assumption is false
 
#### Proof by contradiction
+ Assume a station $k \neq N_s, 0 \leqslant k \leqslant N_s$, couldn't reach if starting from $N_s$

+ Therefore, the assumption --- exist a station $k \neq N_s, 0 \leqslant k \leqslant N_s$, couldn't reach if starting from $N_s$ --- must be false
+ $N_s$ is the unique answer to the problem
