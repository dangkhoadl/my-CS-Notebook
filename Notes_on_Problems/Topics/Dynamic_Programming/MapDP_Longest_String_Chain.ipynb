## [Longest String Chain](https://leetcode.com/problems/longest-string-chain/)

- **Predecessor**: word1 is a predecessor of word2 if and only if 
    + we can add exactly one letter anywhere in word1 to make it equal to word2.
    + For example, "abc" is a predecessor of "abac".
- **Word chain**: a sequence of words [word_1, word_2, ..., word_k]
    + word_1 is a predecessor of word_2
    + word_2 is a predecessor of word_3
    + ...
- Give a list of strings. Find the longest string chain

```
Input: ["a","b","ba","bca","bda","bdca"]
Output: 4
Explanation: one of the longest word chain is "a","ba","bda","bdca".
```

- Notes:
    + 1 <= words.length <= 1000
    + 1 <= words[i].length <= 16
    + words[i] only consists of English lowercase letters.

#### Solution BottomUp DP

```C++
class Solution {
public:
    int longestStrChain(vector<string>& words) {
        /*------------ Normal sort ------------*/
        // sort(words.begin(), words.end(), 
        //     [](string &a, string &b){return a.size()<b.size();});

        /*------------ Count sort ------------*/
        // Count 
        vector<vector<string>> cnt(17);
        for(auto &w:words) {
            cnt[w.size()].push_back(w);
        }

        // Sort
        words.clear();
        for(auto &vec:cnt) {
            if(vec.size() == 0) continue;
            for(auto &w:vec) words.push_back(w);
        }

        /*------------ DP ------------*/
        unordered_map<string, int> dp;
        for(auto &w:words) {
            if(dp.count(w) != NULL) continue;

            for(int i=0; i<w.size(); ++i) {
                string predecessor = w.substr(0,i) + w.substr(i+1);
                dp[w] = max(dp[w], dp[predecessor]+1);
            }
        }

        /*------------ Relax ------------*/
        int ans = 0;
        for(auto e:dp) ans = max(ans, e.second);
        return ans;
    }
};
```
