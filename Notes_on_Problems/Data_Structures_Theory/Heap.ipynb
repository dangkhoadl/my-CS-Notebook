```C++
class MinHeap {
private:
    vector<int> _H;
    int _N;
private:
    int __get_parent(int i) {
        return (i-1) / 2;
    }
    int __get_left_child(int i) {
        return 2*i + 1;
    }
    int __get_right_child(int i) {
        return 2*i + 2;
    }

    void __sift_up(int i) {
        // Root
        if(i <= 0) return;

        // If parent >= _H[i] swap and continue
        int p = __get_parent(i);
        if(_H[i] < _H[p]) {
            swap(_H[i], _H[p]);
            __sift_up(p);
        }
    }
    void __sift_down(int i) {
        // Get left, right child index
        int l = __get_left_child(i);
        int r = __get_right_child(i);

        // Find the min node (cur/left_child/right_child)
        int min_index = i;
        if(l<_N && _H[l] < _H[min_index]) min_index = l;
        if(r<_N && _H[r] < _H[min_index]) min_index = r;

        // If 1 child is smaller swap and continue
        if(min_index != i) {
            swap(_H[i], _H[min_index]);
            __sift_down(min_index);
        }
    }
public:
    /* --- Min Priority Queue --- */
    MinHeap() : _H(), _N(0) {}
    int size() {
        return _N;
    }
    bool empty() {
        if(_N == 0) return true;
        return false;
    }
    int top() {
        return _H[0];
    }
    void pop() {
        // Erase root and Move back_element to root
        // Heapify = __sift_down(root)
        _H[0] = _H[_N-1];
        __sift_down(0);

        // remove back and resize
        _H.pop_back();
        _N -= 1;
    }
    void push(int x) {
        // Insert back
        // Heapify = __sift_up(back_element)
        _H.push_back(x);
        __sift_up(_N);
        _N += 1;
    }

    /* --- In-place heapsort (decreasing)--- */
    MinHeap(const vector<int> &A) : _H(A), _N(A.size()) {
        // build heap O(n)
        for(int i=_N/2; i>=0; --i) __sift_down(i);
        prit(_H)
    }
    vector<int> sort() {
        while(_N--) {
            swap(_H[0], _H[_N]);
            __sift_down(0);
        }
        return _H;
    }
};
```
