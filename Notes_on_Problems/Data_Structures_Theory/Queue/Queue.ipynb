## Queue Doubly LinkedList

```C++
class Queue {
private:
    struct Node {
        int val;
        Node *next, *prev;
        Node(int x) : val(x), next(NULL), prev(NULL) {}
    };

    // pointers
    Node *head_, *tail_;
public:
    Queue() : head_(NULL), tail_(NULL) {}

    int back() {
        return tail_ != NULL ? tail_->val : NULL;
    }
    void push_back(int val) {
        // Create new node
        Node *newNode = new Node(val);

        // Case empty linkedlist
        if(head_ == NULL) {
            head_ = tail_ = newNode;
            return;
        }

        // Case not empty
        tail_->next = newNode;
        newNode->prev = tail_;

        tail_ = newNode;
    }
    void pop_back() {
        // Case empty linkedlist
        if(tail_ == NULL) return;

        // Point delete element at tail
        Node *temp = tail_;

        // Case 1 element left
        if(head_ == tail_) {
            tail_ = head_ = NULL;
        }

        // Case >= 2 elements left
        else {
            tail_ = tail_->prev;
            tail_->next = NULL;
        }
        
        // delete the Node
        delete temp;
    }

    int front() {
        return head_ != NULL ? head_->val : NULL;
    }
    void push_front(int val) {
        // Create new node
        Node *newNode = new Node(val);

        // Case empty linkedlist
        if(head_ == NULL) {
            head_ = tail_ = newNode;
            return;
        }

        // Case not empty
        newNode->next = head_;
        head_->prev = newNode;

        head_ = newNode;
    }
    void pop_front() {
        // Case empty linkedlist
        if(head_ == NULL) return;

        // Point delete element at head
        Node *temp = head_;

        // Case 1 element left
        if(head_ == tail_) {
            tail_ = head_ = NULL;
        }

        // Case >= 2 elements left
        else {
            head_ = head_->next;
            head_->prev = NULL;
        }

        // delete the Node
        delete temp;
    }
};
```
