## [Trapping Rain Water 1D](https://leetcode.com/problems/trapping-rain-water/)
- Given n non-negative integers representing an elevation map where the width of each bar is 1, compute how much water it is able to trap after raining.

<img src="./img/1.png" alt="drawing" width="650"/>

- Example
```
Input: [0,1,0,2,1,0,1,3,2,1,2,1]
Output: 6
```

#### Solution 1
###### Observations
- Trap_water[i] hold by peak to the right and peak to the left
- Amount of water[i] = min(right_peak, left_peak) - height[i]
- Rear cell(0, N-1) cant hold water

```C++
class Solution {
public:
    int trap(vector<int>& height) {
        int N = height.size();
        if(N==0) return 0;
        int runner;
        int water;
        
        // Cache peak left
        vector<int> peak_left(N, 0);
        runner = 0;
        for(int i=0; i<N; ++i) {
            peak_left[i] = runner = max(height[i], runner);
        }

        // Cache peak right
        vector<int> peak_right(N, 0);
        runner = 0;
        for(int i=N-1; i>=0; --i) {
            peak_right[i] = runner = max(height[i], runner);
        }

        // Find the trapping water
        water = 0;
        for(int i=1; i<N-1; ++i) {
            if(height[i] < peak_left[i-1] && height[i] < peak_right[i+1]) {
                water += min(peak_left[i-1], peak_right[i+1]) - height[i];
            }
        }

        return water;
    }
};
```

#### Solution 2
###### Observations
- Water tends to be hold by the minimum col from the perimeters
- As the water trapped at cell[i_] -> cell[i_] become the new wall height =cell[i] 

```C++
class Solution {
public:
    typedef pair<int,int> pii;
    vector<int> dir = {-1, 1};
    int trap(const vector<int>& height) {
        int n = height.size();
        if(n==0) return 0;

        // Init visited and Min heap
        vector<bool> visited(n, false);
        priority_queue<pii, vector<pii>, greater<pii>> PQ;

        // Push heap the perimeter 
        PQ.push( {height[0], 0} );
        visited[0] = true;
        PQ.push( {height[n-1], n-1} );
        visited[n-1] = true;

        int i, height_i, i_;
        int water = 0;
        while(!PQ.empty()) {
            height_i = PQ.top().first;
            i = PQ.top().second;
            PQ.pop();

            for(int k=0; k<2; ++k) {
                i_ = i + dir[k];
                if(i_<0 || i_>=n || visited[i_])
                    continue;
                visited[i_] = true;

                // water trapped at cell[i_] -> cell[i_] is the wall height = cell[i]
                if(height_i > height[i_]) {
                    water += (height_i - height[i_]);
                    PQ.push( {height_i, i_} );
                }

                // water leaked at cell[i_] -> cell[i_] is the wall
                else {
                    PQ.push( {height[i_], i_});
                }
            }
        }
        return water;
    }
};
```

## [Trapping Rain Water 2D](https://leetcode.com/problems/trapping-rain-water-ii/)

#### Solution
###### Obervations

```
12 13 1  12            12 13 1  12
13 4  13 12       ->   13 x  13 12
13 8  10 12            13 x  x  12
12 13 12 12            12 13 12 12
13 13 13 13            13 13 13 13
```
- Waters trapped by the cell 12 from the perimeter


```C++
class Solution {
public:
    typedef pair<int,int> pii;
    vector<pii> dir = {
            {0, 1},
            {0, -1},
            {1, 0},
            {-1, 0}};
    int trapRainWater(vector<vector<int>>& heightMap) {
        int n = heightMap.size();
        if(n == 0) return 0;
        int m = heightMap[0].size();

        // visited
        vector<vector<bool>> visited(n, vector<bool>(m, false));

        // Min heap
        priority_queue<pii, vector<pii>, greater<pii>> PQ;

        // Push heap the perimeter
        for(int i=0; i<n; ++i) for(int j=0; j<m; ++j) {
            if( (i==0 || i==n-1 || j==0 || j==m-1) && !visited[i][j]) {
                PQ.push( {heightMap[i][j], i*m+j} );
                visited[i][j] = true;
            }
        }

        // leaking simulation
        int water = 0, heightMap_ij;
        int i,i_, j,j_;
        while(!PQ.empty()) {
            heightMap_ij = PQ.top().first;
            i = PQ.top().second / m;
            j = PQ.top().second % m;
            PQ.pop();

            for(int k=0; k<4; ++k) {
                i_ = i + dir[k].first;
                j_ = j + dir[k].second;

                if(i_<0 || i_>=n || j_<0 || j_>= m || visited[i_][j_])
                    continue;
                visited[i_][j_] = true;

                // water trapped at cell[i_][j_] -> cell[i_][j_] is the wall, height = cell[i][j]
                if(heightMap_ij > heightMap[i_][j_]) {
                    water += (heightMap_ij - heightMap[i_][j_]);
                    PQ.push( {heightMap_ij, i_*m + j_} );
                }

                // water leaked at cell[i_][j_] -> cell[i_][j_] is the wall
                else {
                    PQ.push( {heightMap[i_][j_], i_*m + j_} );
                }
            }
        }
        return water;
    }
};
```
