
## All Permutations of a string 1
- Find n! permutation of a string
- Example s = "ABC":

```
ABC
ACB
BAC
BCA
CBA
CAB
```


#### Solution
- dfs the tree

<img src="./img/5.jpg" alt="drawing" width="650"/>


```C++
string S = "ABC";

string permu;
unordered_set<char> vis;
void dfs(int idx) {
    if(idx == S.size()) {
        cout << permu << endl;
        return;
    }

    for(char &c:S) {
        if(vis.count(c) == 0) {
            permu += c;
            vis.insert(c);
            
            dfs(idx+1);
            
            vis.erase(c);
            permu.pop_back();
        }
    }
}
```

## All Permutations of a string 2
- Find n! permutation of a string
- Example s = "ABC":

```
ABC
ACB
BAC
BCA
CBA
CAB
```

#### Solution
- dfs the tree
<img src="./img/1.jpg" alt="drawing" width="800"/>



```C++
string s = "ABC";
int n = s.size();

// f(0)
void f(int x) {
    if(x == n) {
        cout << s << endl;
        return;
    }

    for(int c=x; c<n; ++c) {
        swap(s[x], s[c]);

        f(x+1);
        
        // Backtrack
        swap(s[x], s[c]);
    }
}
```
