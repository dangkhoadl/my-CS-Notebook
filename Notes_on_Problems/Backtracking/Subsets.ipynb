## All subsets of n
- Print all $2^n$ subsets of the set: (0,1,2,...n-1)
- Example:

```
n = 2
    {}
    1 
    0 
    0 1
    
n = 3
    {}
    2 
    1 
    1 2 
    0 
    0 2 
    0 1 
    0 1 2 
```

## Solution
- dfs the tree

<img src="./img/4.jpg" alt="drawing" width="800"/>



```C++
int n = 3;

// f(0)
vector<int> subset;
void f(int x) {
    if(x == n) {
        // print subset
        for(auto &sub: subset)
            cout << sub << ' ';
        cout << endl;
        return;
    }

    // Add x
    subset.push_back(x);
    f(x+1);
    subset.pop_back();

    // Not add x
    f(x+1);
}
```

#### Template 2 - inplace

```C++
int n = 3;

// f(0, {})
void f(int x, const vector<int> &subset) {
    if(x == n) {
        // print subset
        for(auto &sub: subset)
            cout << sub << ' ';
        cout << endl;
        return;
    }

    // Add x
    auto new_set = subset; new_set.push_back(x);
    f(x+1, new_set);
    
    // Not add x
    f(x+1, subset);
}
```

#### Template 3

```C++
int n = 3;

// f(0)
vector<int> subset;
void f(int x) {
    // back track
    for(int i=x; i<n; ++i) {
        subset.push_back(i);
        f(i+1);
        subset.pop_back();
    }

    // x == n
    for(auto &sub: subset)   // print subset
        cout << sub << ' ';
    cout << endl;
}
```
