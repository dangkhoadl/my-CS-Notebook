## [Number of Squareful Arrays](https://leetcode.com/problems/number-of-squareful-arrays/)
- An array is squareful if for every pair of adjacent elements (A[i] and A[i+1]), their sum is a perfect square.
- Return the number of permutations of A that are squareful. All permutations must be unique
- **Examples**

```
Input: [1,17,8]
Output: 2
Explanation: 
    [1,8,17] and [17,8,1] are the valid permutations
    
Input: [2,2,2]
Output: 1
```

#### Solution

```C++
class Solution {
public:
    int N;
    int cnt;
    vector<int> A;
    vector<bool> chosen;

    bool is_sqrt(int x) {
        int v = sqrt(x);
        return v*v == x;
    }
    void dfs(int x, int cur) {
        if(x == N) {
            ++cnt;
            return;
        }

        for(int c=0; c<N; ++c) {
            if(chosen[c]) continue;
            if(c>0 && A[c] == A[c-1] && !chosen[c-1]) continue; // skip duplicates

            if(cur == -1 || is_sqrt(cur + A[c])) {
                chosen[c] = true;
                dfs(x+1, A[c]);
                chosen[c] = false;
            }
        }
    }
    int numSquarefulPerms(vector<int>& nums) {
        A = nums;
        N = A.size();
        chosen.assign(N, false);

        sort(A.begin(), A.end());  // skip duplicates
        cnt=0;
        dfs(0, -1);
        return cnt;
    }
};

```
