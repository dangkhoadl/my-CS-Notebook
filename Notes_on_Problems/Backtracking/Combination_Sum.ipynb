# Combination Sum
- Given a set of candidate numbers (candidates)
- and a target number (target)
- find all unique combinations in candidates where the candidate numbers sums to target
- **The solution set must not contain duplicate combinations.**

## [Combination Sum - 1 element can be used multiple times](https://leetcode.com/problems/combination-sum/)
- Given a set of candidate numbers (candidates) **without duplicates** and a target number (target)
- find all unique combinations in candidates where the candidate numbers sums to target.
    + The **same repeated number** may be chosen from candidates unlimited number of times.

- Example

```
Input: candidates = [2,3,6,7], target = 7,
A solution set is:
[
  [7],
  [2,2,3]
]

Input: candidates = [2,3,5], target = 8,
A solution set is:
[
  [2,2,2,2],
  [2,3,3],
  [3,5]
]
```

#### Solution

```C++
class Solution {
public:
    int N;
    vector<int> a;

    vector<vector<int>> ans;
    vector<int> res;
    void dfs(int x, int remain) {
        if(remain < 0) return;
        if(remain == 0) {
            ans.push_back(res);
            return;
        }

        for(int i=x; i<N; ++i) {    // Start from x as we dont consider permutation
            if(remain - a[i] < 0) continue;

            res.push_back(a[i]);
            dfs(i, remain-a[i]);    // 1 element can be used multiple times
            res.pop_back();
        }
    }

    vector<vector<int>> combinationSum(vector<int>& candidates, int target) {
        a = candidates;
        N = a.size();

        dfs(0, target);
        return ans;
    }
};
```

## [Combination Sum - 1 element can be used only once](https://leetcode.com/problems/combination-sum-ii/)
- Given a collection of candidate numbers (candidates) and a target number (target), 
- find all unique combinations in candidates where the candidate numbers sums to target.
    + Each number in candidates may only be **used once** in the combination.
- Example

```
Input: candidates = [10,1,2,7,6,1,5], target = 8,
A solution set is:
[
  [1, 7],
  [1, 2, 5],
  [2, 6],
  [1, 1, 6]
]

Input: candidates = [2,5,2,1,2], target = 5,
A solution set is:
[
  [1,2,2],
  [5]
]
```

#### Solution

```C++
class Solution {
public:
    int N;
    vector<int> a;

    vector<vector<int>> ans;
    vector<int> res;
    void dfs(int x, int remain) {
        if(remain < 0) return;
        if(remain == 0) {
            ans.push_back(res);
            return;
        }

        for(int i=x; i<N; ++i) {                   // Start from x as we dont consider permutation
            if(remain - a[i] < 0) continue;
            if (i > x && a[i] == a[i-1]) continue; // skip duplicates

            res.push_back(a[i]);
            dfs(i+1, remain-a[i]); // 1 element used once
            res.pop_back();
        }
    }

    vector<vector<int>> combinationSum2(vector<int>& candidates, int target) {
        a = candidates;
        N = a.size();

        sort(a.begin(), a.end()); // skip duplicates
        dfs(0, target);
        return ans;
    }
};
```
