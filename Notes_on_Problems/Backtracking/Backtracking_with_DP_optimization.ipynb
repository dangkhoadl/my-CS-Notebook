# Problem
- Given MxN grid
- Find a path go through N rows with minimum cost
+ Constraint: Only 3 moves available
    + Down left
    + Down
    + Down Right
+ Row 0: Can choose any cell to start

#### Example
- Input

```
5 5
1  3  1  2  6
10 2  5  4  15
10 9  6  7  1
2  7  1  5  3
8  2  6  1  9
```

- Output: 10

- Explanation

```
1   3  [1]  2    6
10  2   5  [4]   15
10  9   6   7   [1]
2   7   1   5   [3]
8   2   6  [1]   9
```

# Solution - backtrack

```Cpp
int A[103][103];


int ans, res;
void get(int i, int pre) {
    if(i<0 || i>N || pre<-1 || pre>=M) return;
    if(i == N) {
        ans = min(ans, res);
        return;
    }

    // i = 0
    if(pre == -1) {
        for(int j=0; j<M; ++j) {
            res += A[i][j];
            get(i+1, j);
            res -= A[i][j];
        }
        return;
    }

    // else
    for(int j=pre-1; j<=pre+1; ++j) {
        if(j<0 || j>=M) continue;
        res += A[i][j];
        get(i+1, j);
        res -= A[i][j];
    }
}


void sol() {
    res = 0, ans = INT_MAX;
    get(0, -1);
    cout << ans << endl;
}
```

# Solution - backtrack with DP optimization

```cpp
int A[103][103];


int ans;
int dp[103][103];
void get(int i, int pre) {
    if(i<0 || i>N || pre<-1 || pre>=M) return;
    if(i == N) {
        ans = min(ans, dp[i][pre]);
        return;
    }

    // i = 0
    if(pre == -1) {
        for(int j=0; j<M; ++j) {
            if(dp[i+1][j] > dp[i][pre] + A[i][j]) {
                dp[i+1][j] = dp[i][pre] + A[i][j];
                get(i+1, j);
            }
        }
        return;
    }

    // else
    for(int j=pre-1; j<=pre+1; ++j) {
        if(j<0 || j>=M) continue;

        if(dp[i+1][j] > dp[i][pre] + A[i][j]) {
            dp[i+1][j] = dp[i][pre] + A[i][j];
            get(i+1, j);
        }
    }
}



void sol() {
    ans = INT_MAX;
    for(int i=0; i<=N; ++i) for(int j=0; j<=M; ++j) {
        dp[i][j] = INT_MAX;
    }
    get(0, -1);

    cout << ans << endl;
}
```
