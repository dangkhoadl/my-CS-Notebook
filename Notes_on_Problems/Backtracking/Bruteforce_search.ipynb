# [24 Game](https://leetcode.com/problems/24-game/)
- **Given**
    + Array `cards` length = 4
        + `cards[i]` in range `[1, 9]`
    + Operations: `['+', '-', '*', '/']`, `'('` and `')'`
    + Arrange the numbers on these cards in a mathematical expression. Eg: `[1,2,3,4]` -> `4 / (1 - 2 / 3)` = 12
- Return `true` if can get such expression that evaluates to **24**

#### Example 1

```
Input: cards = [4,1,8,7]
Output: true
Explanation: (8-4) * (7-1) = 24
```

#### Example 2

```
Input: cards = [1,2,1,2]
Output: false
```

# Solution
- Bruteforce search

```C++
class Solution {
public:
    bool is_eq(double a, double b) {
        return fabs(a - b) < 1e-8;
    }
    bool dfs(const vector<double> &cur) {
        if(cur.size() == 1) return is_eq(cur[0], 24.0);
        for(int i=0; i+1<cur.size(); ++i) for(int j=i+1; j<cur.size(); ++j) {
            vector<double> next_;
            for(int x=0; x<cur.size(); ++x) if(x!=i && x!=j) next_.push_back(cur[x]);

            // +
            next_.push_back(cur[i] + cur[j]);
            if(dfs(next_)) return true;

            // -
            next_.back() = cur[i] - cur[j];
            if(dfs(next_)) return true;

            next_.back() = cur[j] - cur[i];
            if(dfs(next_)) return true;

            // *
            next_.back() = cur[i] * cur[j];
            if(dfs(next_)) return true;

            // /
            if(!is_eq(cur[j], 0.0)) {
                next_.back() = cur[i] / cur[j];
                if(dfs(next_)) return true;
            }

            if(!is_eq(cur[i], 0.0)) {
                next_.back() = cur[j] / cur[i];
                if(dfs(next_)) return true;
            }
        }

        return false;
    }
    bool judgePoint24(vector<int>& cards) {
        vector<double> cards_double;
        for(int &x:cards) cards_double.push_back(static_cast<double>(x));
        return dfs(cards_double);
    }
};
````
