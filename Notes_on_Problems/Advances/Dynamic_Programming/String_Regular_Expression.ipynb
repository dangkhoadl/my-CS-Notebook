## [Regular Expression Matching](https://leetcode.com/problems/regular-expression-matching/)
- Give 2 strings s and p. Find if s and p can be matched
    + s could be empty and contains only lowercase letters a-z.
    + p could be empty and contains only lowercase letters a-z, and characters like . or *.
```
'.' Matches any single character.
'*' Matches zero or more of the preceding element.
    Eg: 'c*' can be
        + ''
        + 'c'
        + 'cc'
        + 'ccc...c'
```
- Examples

```
Input:
    s = "aa"
    p = "a"
Output: false

Input:
    s = "aa"
    p = "a*"
Output: true

Input:
    s = "ab"
    p = ".*"
Output: true

Input:
    s = "aab"
    p = "c*a*b"
Output: true

Input:
    s = "mississippi"
    p = "mis*is*p*."
Output: false

Input:
    s = "a"
    p = "ab*"
Output: false
```

## Solution O($n^2$)
<img src="./img/7.jpg" alt="drawing" width="650"/>

```C++
// s,p start at 0
string s,p;

bool dp[103][103];
bool solveDP() {
    int M = s.size();
    int N = p.size();

    // Init dp
    ms(dp,false);
    dp[0][0] = true;
    
    FOR(i,0,M+1) FOR(j,0,N+1) {
        if(dp[i][j] == false)
            continue;

        // Case normal matching - s[i] match p[j] or s[i] match '.'
        if(i<M && j<N &&
                (s[i] == p[j] || p[j] == '.')) {
            dp[i+1][j+1] = true;
        }

        // Case '*' = ''
        if(j+1<N &&
                p[j+1] == '*') {
            dp[i][j+2] = true;
        }

        // Case: '*' = s[i] or s[i]...s[i]
        if(i<M && j+1<N &&
                p[j+1] == '*' && (s[i] == p[j] || p[j] == '.')) {
            // * = s[i]
            dp[i+1][j+2] = true;

            // * = s[i]...s[i]
            dp[i+1][j] = true;
        }
    }
    return dp[M][N];
}
```
