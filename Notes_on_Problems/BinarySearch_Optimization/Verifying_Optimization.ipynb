#  [Split Array Largest Sum](https://leetcode.com/problems/split-array-largest-sum/)
- Split an array into 
    + m non-empty continuous subarrays
    + minimize the largest sum among these m subarrays.
    
#### Example 1

```
[7,2,5,10,8], m = 2

4 ways of spliting
    [7] [2,5,10,8] largest_sum = 25
    [7,2] [5,10,8] largest_sum = 23
    [7,2,5] [10,8] largest_sum = 18
    [7,2,5,10] [8] largest_sum = 24
    
=> minimum largest_sum = 18
```

#### Example 2

```
[1,2,3,4,5], m = 2

=> minimum largest_sum = 9
[1,2,3] [4,5]
```

#### Example 3

```
[1,4,4], m = 3
=> minimum largest_sum = 4
[1] [4] [4]
```

# Solution
#### Solution - O(Nlog(N))
- Range of max_val:

``` 
    0        [step=1]                         total_sum       
    |--|--|--|--|--|--|--|--|--|--|--|--|--|--|--|
```
- m subarray -> m-1 cuts
- for each max_val we can check if the array is splitable with m-1 number of cuts and the largest segment <= max_val-> Find the minimum max_val that splitable 

```
    [      ][  ][ ]
```

```C++
class Solution {
private:
    vector<int> _A;

    bool __can_split(int max_val, int cuts) {
        /*--- Check if nums can be split by max_val (at most) and cuts O(n)---*/

        int acc_sum = 0;
        for(int &x:_A) {
            // oversize
            if(x > max_val) return false;

            // Can add x to the current slice
            if(acc_sum + x <= max_val) {
                acc_sum += x;
            }

            // Cut, create a new slice with x
            else {
                cuts -= 1;
                if(cuts < 0) return false;

                acc_sum = x;
            }
        }
        return true;
    }
public:
    int splitArray(vector<int>& nums, int M) {
        // Init private
        _A = nums;

        // Init search range [0, sum]
        int l = 0;
        int r = 0;
        for(int &x:_A) r += x;

        // Bin search
        while(l<=r) {
            int val = l + (r-l)/2;

            // can_split -> decrease val -> search left
            if(__can_split(val, M-1)) {
                r = val - 1;
            }

            // decrease val -> search right
            else {
                l = val + 1;
            }
        }
        return l;
    }
};
```
