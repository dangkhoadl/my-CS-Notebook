# Problem
- Have N candies
- Each time
    + John takes k candies
    + Alice takes 10% of remaining candies
- If number of candies is less than k: John take all the remaining
- Choose the minimum k that John will take at least half of N candies

#### Constraint
- $N \leq 10^{18}$ 

#### Example
```
N = 68

// Solution
min k = 3

// Explanation
68-3 = 65; 65/10 = 6; 65-6 = 59
59-3 = 56; 56/10 = 5; 56-5 = 51
51-3 = 48; 48/10 = 4; 48-4 = 44
44-3 = 41; 41/10 = 4; 41-4 = 37
…..
…..
…..
6-3 = 3; 3/10 = 0; 3-0 = 3
3-3 = 0; 0/10 = 0; 0-0 = 0
```

#### Solution

```cpp
long N;
bool check(long N, long k) {
    long john = 0;
    long candies = N;
    while(true) {
        if(candies <= k) {
            john += candies;
            break;
        }

        john += k;
        candies -= k;
        candies -= candies/10;
    }
    if(john*2 >= N) return true;
    return false;
}


long sol() {
    long l = 1, r = N;
    long ans=0;
    while(l<=r) {
        long m = l + (r-l)/2;
        if(check(N, m)) {
            r = m - 1;
            ans = m;
        } else {
            l = m + 1;
        }
    }
    return ans;
}
```
