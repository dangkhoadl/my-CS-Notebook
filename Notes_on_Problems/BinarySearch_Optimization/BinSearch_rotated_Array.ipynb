
# [Search in Rotated Sorted Array](https://leetcode.com/problems/search-in-rotated-sorted-array/)
- Given
    + A sorted array `A` but is roated at a pivot x, x is unknown. For example
        + `[4,5,6,7,0,1,2]`
    + A `target`
- return the index of `target` if it is in `A`
    + -1 if it is not in `A`

#### Examples

```
Example 1:

Input: nums = [4,5,6,7,0,1,2], target = 0
Output: 4

Example 2:
Input: nums = [4,5,6,7,0,1,2], target = 3
Output: -1

Example 3:
Input: nums = [1], target = 0
Output: -1
```


#### Solution
- The array values would be at this shape -> Divide and conquer

<img src="./img/1.png" width="400"/>

- If we choose a pivot `p`. There are 3 cases that we would search on the left hand side (index in between [l,p)

<img src="./img/2.png" width="900"/>

- Otherwise, search on the right hand side

#### Code

```C++
class Solution {
public:
    int search(vector<int>& A, int target) {
        int l = 0;
        int r = A.size()-1;
        while (l <= r) {
            int p = l + (r-l) / 2;
            if(A[p] == target) return p;

            // Search left
            if( (A[0] <= target && target < A[p]) || (target < A[p] && A[p] < A[0]) || (A[p] < A[0] && A[0] <= target))
                r = p - 1;
            // Search right
            else
                l = p + 1;
        }

        return -1;
    }
};
```

# [Find Minimum in Rotated Sorted Array](https://leetcode.com/problems/find-minimum-in-rotated-sorted-array/)

- Given
    + A sorted array `A` but is roated at a pivot x, x is unknown. For example
        + `[4,5,6,7,0,1,2]`
        + `[0,1,2,4,5,6,7]`
- Return: the minimum element of this array

#### Examples

```
Input: nums = [3,4,5,1,2]
Output: 1

Input: nums = [4,5,6,7,0,1,2]
Output: 0

Input: nums = [11,13,15,17]
Output: 11
```

#### Solution - Bin search
- **Idea**: Reduce the problem to "Find the leftmost elem that less than A[0]" 
<img src="./img/1.png" width="400"/>

```C++
class Solution {
public:
    int findMin(vector<int> &A) {
        int l = 0;
        int r = A.size()-1;

        // Edge case: Check if Array not rotated
        if(A[0] <= A[r]) return A[0]; 

        while(l<r) {
            int m = l + (r-l)/2;

            if(A[m] >= A[0]) {
                // search right
                l = m + 1;
            } else {
                // search left
                r = m;
            }
        }
        return A[l];
    }
};
```
