## [Remove Duplicate Letters](https://leetcode.com/problems/remove-duplicate-letters/)
- **Given**
    + a string `S`: contains only lowercase letters
- **Return**:
    + Remove duplicate letters so that every letter appear once and only once.
    + Return string is the **smallest in lexicographical order** among all possible results.

#### Examples

```
Input: "bcabc"
Output: "abc"

Input: "cbacdcbc"
Output: "acdb"
```

## Solution 1 - Monotonic Stack
- Build a **Monotonic increasing stack** as possible, **Greedily** removes back chars/digits that > adding chars/digits while maintaining all chars exists

#### Code

```Cpp
class Solution {
public:
    string removeDuplicateLetters(const string &S) {
        // Build Last index occur
        unordered_map<char, int> last_idx;
        for(int i=0; i<S.size(); ++i) {
            last_idx[S[i]] = i;
        }

        // min <<< max [back] S[i]
        string ans = "";
        unordered_set<char> vis;
        for(int i=0; i<S.size(); ++i) {
            if(vis.count(S[i]) == 1) continue;

            // Pop stack
            while(!ans.empty() && ans.back() > S[i] && last_idx[ans.back()] > i) {
                vis.erase(ans.back());
                ans.pop_back();
            }

            // Push stack
            ans += S[i];
            vis.insert(S[i]);
        }
        return ans;
    }
};
```

## Solution 2 - Recursive

- Detect the smallest lecographical char possible without missing remaining char
- append to the new string and Remove all
- Example: cbacdcbc  
get(cb`a`cdcbc)  
= a + get(`c`dcbc)  
= ac + get(`d`b)  
= acd + get(`b`)  
= acdb

#### Code

```C++
class Solution {
private:
    string get(const string &S) {
        if(S.size() == 0) return "";

        // Build count
        unordered_map<char, int> cnt;
        for(const char &c: S) cnt[c] += 1;

        // Greedy select S[p]: the smallest lexicographical and all character still exist
        int p = 0;
        for(int i=0; i<S.size(); ++i) {
            if(S[i] < S[p]) p = i;

            cnt[S[i]] -= 1;
            if(cnt[S[i]] == 0) break;
        }

        // Construct S_ = S(p+1) remove all S[p]
        string S_ = "";
        for(const char &c: S.substr(p+1)) {
            if(c != S[p]) S_ += c;
        }

        // Return: S[p] + get(S_)
        return S[p] + get(S_);
    }
public:
    string removeDuplicateLetters(const string &s) {
        return get(s);
    }
};
```
