# [Next Greater Element II](https://leetcode.com/problems/next-greater-element-ii/)

- **Given**
    + Array `A` size N
- **Return**: The next greater number for every element in `A`
    + If not exist, return `-1`
    + you could search circularly to find its next greater number
    
    
#### Examples

```
Input: A = [1,2,1]
Output: [2,-1,2]
```

```
Input: A = [1,2,3,4,3]
Output: [2,3,4,-1,4]
```

## Solution
- Maintain a **Monotonic decreasing stack**, If meet a number > stack top, pop the stack to update 

#### Code

```Cpp
class Solution {
public:
    vector<int> nextGreaterElements(vector<int> &A) {
        int N = A.size();
        vector<int> B(A);
        for(auto num: A) B.push_back(num);

        vector<int> ans(N, -1);
        
        // max >>> min [back] i
        vector<int> stck;
        for(int i=0; i<B.size(); ++i) {
            while(!stck.empty() && B[stck.back()] < B[i]) {
                ans[stck.back()%N] = A[i%N];
                stck.pop_back();
            }

            stck.push_back(i);
        }
        return ans;
    }
};
```

# [Number of Visible People in a Queue](https://leetcode.com/problems/number-of-visible-people-in-a-queue/)

- **Given**
    + Array `H`: heights of n people
- **Return**: Number of people each person can see. Assume person `i` can see person `j`
    + `i` < `j`: j on the right hand side of i
    + min(H[i], H[j]) > max(H[i+1], H[i+2], ..., H[j-1]): No person in between `i` and `j` blocks the view

#### Examples

<img src="./assets/1.jpg" width="400"/>


```
Input: H = [10,6,8,5,11,9]
Output: [3,1,2,1,1,0]
Explanation:
    Person 0 can see person 1, 2, and 4.
    Person 1 can see person 2.
    Person 2 can see person 3 and 4.
    Person 3 can see person 4.
    Person 4 can see person 5.
    Person 5 can see no one since nobody is to the right of them.
    
    
Input: H = [5,1,2,3,10]
Output: [4,1,1,1,0]
```

## Solution
- Maintain a **Monotonic decreasing stack** as possible, while a person taller than stack top, pop and process the top

#### Code

```Cpp
class Solution {
public:
    vector<int> canSeePersonsCount(vector<int> &H) {
        int N = H.size();
        vector<int> ans(N, 0);

        // max >>> min [back] j
        vector<int> stck;
        for(int j=0; j<N; ++j) {
            // pop i can see j
            while(!stck.empty() && H[stck.back()] < H[j]) {
                int i = stck.back();
                ans[i] += 1;
                stck.pop_back();
            }

            // Most left i
            if(!stck.empty()) {
                int i = stck.back();
                ans[i] += 1; 
            }

            // Add stack
            stck.push_back(j);
        }

        return ans;
    }
};
```
