# [3-Sum](https://leetcode.com/problems/3sum/)

- **Given**
    + Array `A`
- **Find** all the triplets `[nums[i], nums[j], nums[k]]` such that
    + `i != j, i != k, and j != k`
    + `nums[i] + nums[j] + nums[k] == 0`

#### Examples

```
Input: nums = [-1,0,1,2,-1,-4]
Output: [[-1,-1,2],[-1,0,1]]
```

#### Solution - $O(N^2)$
- Reduce to 2 pointers
    + Have 3 pointers to scan: `l,m,r`
    + Fixed `m` -> slide `l`,`r`
    + Ensure `l < m < r`

```Cpp
class Solution {
public:
    vector<vector<int>> threeSum(vector<int> &A) {
        if(A.size() < 3) return vector<vector<int>>({});

        // sort inc
        sort(A.begin(), A.end());
        vector<vector<int>> pos;

        // sliding windows
        for(int m=1; m<A.size()-1; ++m) {
            int l = 0;
            int r = A.size() - 1;
            while(l < m && m < r) {
                int three_sum = A[l] + A[m] + A[r];

                if(three_sum == 0) {
                    pos.push_back( {l,m,r} );
                    r -= 1;

                    // Special case
                    if(A[l] == 0 && A[m] == 0 && A[r] == 0) break;
                } else if(three_sum < 0) {
                    l += 1;
                } else if(three_sum > 0) {
                    r -= 1;
                }
            }
        }

        // get ans
        set<vector<int>> ans;
        for(auto &cand: pos) {
            ans.insert( {A[cand[0]], A[cand[1]], A[cand[2]]} );
        }
        return vector<vector<int>>(ans.begin(), ans.end());
    }
};
```

# [4-sum](https://leetcode.com/problems/4sum/)

- **Given**
    + Array `A`
    + `target`
- **Find** all the quadruplets `[[nums[a], nums[b], nums[c], nums[d]]` such that
    + a, b, c, and d are **distinct**
    + `nums[a] + nums[b] + nums[c] + nums[d] == target`

#### Examples

```
Input: nums = [1,0,-1,0,-2,2], target = 0
Output: [[-2,-1,1,2],[-2,0,0,2],[-1,0,0,1]]
```

```
Input: nums = [2,2,2,2,2], target = 8
Output: [[2,2,2,2]]
```

#### Solution - $O(N^3)$
- Reduce to 2 pointers
    + Have 4 pointers to scan: `l,m1,,m2,r`
    + Fixed `m1`, `m2` -> slide `l`,`r`
    + Ensure `l < m1 < m2 < r`

```Cpp
class Solution {
private:
    long long __cast(int x) {
        return static_cast<long long>(x);
    }
public:
    vector<vector<int>> fourSum(vector<int> &A, int target) {
        if(A.size() < 4) return vector<vector<int>>({});
        vector<vector<int>> pos;

        // sort A inc
        sort(A.begin(), A.end());

        for(int m1=1; m1<A.size()-2; ++m1) {
            for(int m2=m1+1; m2<A.size()-1; ++m2) {
                int l = 0;
                int r = A.size() - 1;
                while(l < m1 && m2 < r) {
                    long long four_sum = __cast(A[l]) + __cast(A[m1]) + __cast(A[m2]) + __cast(A[r]);
                    if(four_sum == target) {
                        pos.push_back( {l,m1,m2,r} );
                        r -= 1;
                    } else if(four_sum < target) {
                        l += 1;
                    } else if(four_sum > target) {
                        r -= 1;
                    }
                }
            }
        }

        // Get ans
        set<vector<int>> ans;
        for(auto &cand: pos) {
            ans.insert( {A[cand[0]], A[cand[1]], A[cand[2]], A[cand[3]]} );
        }
        return vector<vector<int>>(ans.begin(), ans.end());
    }
};
```
