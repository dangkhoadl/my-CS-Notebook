# [1. Longest Substring Without Repeating Characters](https://leetcode.com/problems/longest-substring-without-repeating-characters/)
Given a string, find the length of the longest substring without repeating characters.
- Example 1

```
Input: "abcabcbb"
Output: 3 
Explanation: The answer is "abc", with the length of 3. 
```

- Example 2

```
Input: "bbbbb"
Output: 1
Explanation: The answer is "b", with the length of 1.
```

- Example 3

```
Input: "pwwkew"
Output: 3
Explanation: The answer is "wke", with the length of 3. 
             Note that the answer must be a substring, "pwke" is a subsequence and not a substring.
```

#### Solution


```C++
class Solution {
public:
    int lengthOfLongestSubstring(const string &s) {
        int ans = 0;
        int l = 0;

        unordered_multiset<char> sliding_window;
        for(int r=0; r<s.size(); ++r) {
            // Shrink left if s[r] in window
            while(sliding_window.count(s[r]) != NULL) {
                sliding_window.erase(s[l]);
                l += 1;
            }

            // Add right
            sliding_window.insert(s[r]);

            // Relax
            ans = max(ans, (int)sliding_window.size());
        }
        return ans;
    }
};
```

#### Solution - Compact

```C++
class Solution {
public:
    int lengthOfLongestSubstring(const string &s) {
        int ans = 0;
        int l = 0;

        unordered_map<char, int> sliding_window;
        for(int r=0; r<s.size(); ++r) {
            // Shrink left if s[r] in window
            while(sliding_window.count(s[r]) != NULL) {
                sliding_window[s[l]] -= 1;
                if(sliding_window[s[l]] == 0) sliding_window.erase(s[l]);
                l += 1;
            }

            // Add right
            sliding_window[s[r]] += 1;

            // Relax
            ans = max(ans, r-l+1);
        }
        return ans;
    }
};
```

# [2. Fruit Into Baskets](https://leetcode.com/problems/fruit-into-baskets/)

```
In a row of trees, the i-th tree produces fruit with type tree[i].

You start at any tree of your choice, then repeatedly perform the following steps:

    1. Add one piece of fruit from this tree to your baskets.  If you cannot, stop.
    2. Move to the next tree to the right of the current tree.  If there is no tree to the right, stop.

Note that you do not have any choice after the initial choice of starting tree: you must perform step 1, then step 2, then back to step 1, then step 2, and so on until you stop.

You have two baskets, and each basket can carry any quantity of fruit, but you want each basket to only carry one type of fruit each.

What is the total amount of fruit you can collect with this procedure?
```

- **Rephrasing**: Find the maximum length subarray that contains up to two distinct integers?

#### Examples

```
Input: [1,2,1]
Output: 3
Explanation: We can collect [1,2,1].

Input: [0,1,2,2]
Output: 3
Explanation: We can collect [1,2,2].
If we started at the first tree, we would only collect [0, 1].

Input: [1,2,3,2,2]
Output: 4
Explanation: We can collect [2,3,2,2].
If we started at the first tree, we would only collect [1, 2].

Input: [3,3,3,1,2,1,1,2,3,3,4]
Output: 5
Explanation: We can collect [1,2,1,1,2].
If we started at the first tree or the eighth tree, we would only collect 4 fruits.
```

#### Solution
```C++
class Solution {
public:
    int totalFruit(vector<int> &A) {
        int l = 0;

        unordered_map<int, int> baskets;
        int total_fruits = 0;

        int ans = 0;
        for(int r=0; r<A.size(); ++r) {
            // Add right
            baskets[A[r]] += 1;
            total_fruits += 1;

            // Shrink left: maintain basket max 2 types of fruit
            while(baskets.size() > 2) {
                baskets[A[l]] -= 1;
                if(baskets[A[l]] == 0) baskets.erase(A[l]);
                l += 1;
                total_fruits -= 1;
            }

            // Relax
            ans = max(ans, total_fruits);
        }
        return ans;
    }
};
```

# [3.Minimum Size Subarray Sum](https://leetcode.com/problems/minimum-size-subarray-sum/)
- Given
    + An array of positive integers `A` and a positive integer `target`
- Find the min length of a contiguous subarray
    + such that the sum is $\geq$ target
    + If there is no such subarray, return 0

#### Examples

```
Input: target = 7, nums = [2,3,1,2,4,3]
Output: 2
Explanation: The subarray [4,3] has the minimal length under the problem constraint.
```

```
Input: target = 4, nums = [1,4,4]
Output: 1
```

```
Input: target = 11, nums = [1,1,1,1,1,1,1,1]
Output: 0
```

#### Solution

```C++
class Solution {
public:
    int minSubArrayLen(int target, vector<int>& A) {
        int l = 0, r = 0;
        int ans = INT_MAX;
        int sum = 0;
        for(int r=0; r<A.size(); ++r) {
            sum += A[r];

            while(sum >= target && l<=r) {
                ans = min(ans, r-l+1);

                sum -= A[l];
                l += 1;
            }
        }
        return (ans == INT_MAX) ? 0: ans;
    }
};
```
