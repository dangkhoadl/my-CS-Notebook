# [1. Increasing Triplet Subsequence - 123 Pattern](https://leetcode.com/problems/increasing-triplet-subsequence)
- **Given**: Array `A`
- **Return**: True if there exists a triple of indices `i < j < k` such that `A[i] < A[j] < A[k]`

#### Examples

```
Input: A = [1,2,3,4,5]
Output: true
Explanation: Any triplet where i < j < k is valid.

Input: A = [5,4,3,2,1]
Output: false

Input: A = [2,1,5,0,4,6]
Output: true
Explanation: [2,4,6], [1,4,6], [0,4,6], [1,5,6], etc.
```

## Solution - Cache
- Cache `min from left` and `max from right`

```Cpp
class Solution {
public:
    bool increasingTriplet(vector<int> &A) {
        int N = A.size();

        // Store A[i]: left min
        vector<int> min_left(N, A[0]);
        for(int i=1; i<N; ++i) min_left[i] = min(min_left[i-1], A[i]);

        // Store A[k]: right max
        vector<int> max_right(N, A[N-1]);
        for(int k=N-2; k>=0; --k) max_right[k] = max(max_right[k+1], A[k]);

        // Relax A[j]
        for(int j=0; j<N; ++j) {
            if(min_left[j] < A[j] && A[j] < max_right[j]) return true;
        }
        return false;
    }
};
```

## Solution - O(1) mem
- Transform problem into **Find the 3rd minumum at i** 

#### Code

```Cpp
class Solution {
public:
    bool increasingTriplet(vector<int>& A) {
        int N = A.size();
        if(N<3) return false;
        
        int min_1 = INT_MAX, min_2 = INT_MAX;
        
        for(int i=0; i<N; ++i) {
            if(A[i] > min_2) return true;
            
            // Relax
            if(min_1 < A[i] && A[i] < min_2) {
                min_2 = A[i];
            } else if(A[i] < min_1) {
                min_1 = A[i];
            }
        }
        return false;
    }
};
```

# [2. 132 Pattern](https://leetcode.com/problems/132-pattern/description/)
- **Given**: Array `A`
- **Return**: True if there exists a triple of indices `i < j < k` such that `A[i] < A[k] < A[j]`

#### Examples

```
Input: A = [1,2,3,4]
Output: false

Input: A = [3,1,4,2]
Output: true
Explanation: [1, 4, 2]

Input: A = [-1,3,2,0]
Output: true
Explanation: [-1, 3, 2], [-1, 3, 0] and [-1, 2, 0]
```

## Solution - Monotonic stack
- Scan from right -> left


#### Code 1
- Store num_j in a **monotonic decreasing stack**
- While poping num_j, update num_k = min(num_js)

```Cpp
class Solution {
public:
    bool find132pattern(vector<int> &A) {
        int N = A.size();
        
        // Store A[k]
        int num_k = INT_MIN;
        // Store A[j]: max >>> min [back]
        vector<int> num_j;

        for(int i=N-1; i>=0; --i) {
            if(A[i] < num_k && !num_j.empty()) return true;

            // Update num_k = min(num_j)
            while(!num_j.empty() && num_j.back() < A[i]) {
                num_k = num_j.back();
                num_j.pop_back();
            }

            // Update num_j
            num_j.push_back(A[i]);
        }
        
        return false;
    }
};
```

#### Code 2

```Cpp
class Solution {
public:
    bool find132pattern(vector<int> &A) {
        int N = A.size();

        // Store A[i]: min form left
        vector<int> min_left(N, 0);
        min_left[0] = A[0];
        for(int i=1; i<N; ++i) min_left[i] = min(min_left[i-1], A[i]);

        //  Store A[k]: max >>> min [back]
        vector<int> num_k;

        // Relax A[j]
        for(int j=N-1; j>=0; --j) {
            int num_i = min_left[j];

            if(A[j] > num_i) {
                // Pop A[k] <= A[i]
                while(!num_k.empty() && num_k.back() <= num_i) num_k.pop_back();

                // Relax
                if(!num_k.empty() && num_k.back() < A[j]) return true;

                // Update num_k
                num_k.push_back(A[j]);
            }
        }
        return false;
    }
};
```
