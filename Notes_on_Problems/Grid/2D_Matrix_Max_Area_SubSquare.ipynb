# [Maximal Square](https://leetcode.com/problems/maximal-square/)

- Given
    + Binary matrix `A`
        + size MxN
        + filled with 0 or 1
- Find the maximum subsquare area
    + Note: subsquare must filled with all 1


# Solution
- DP

```C++
class Solution {
private:
    int __min_3(int a, int b, int c) {
        return min(a, min(b, c));
    }
public:
    int maximalSquare(vector<vector<char>>& A) {
        int M = A.size();
        if(M == 0) return 0;
        int N = A[0].size();

        // dp[i][j]: max sub-square side from (0,0) -> (i,j)
        vector<vector<int>> dp(M, vector<int>(N, 0));

        // Go
        int max_side = 0;
        for(int i=0; i<M; ++i) {
            for(int j=0; j<N; ++j) {
                if(A[i][j] == '1') {
                    if(i==0 || j==0) {
                        dp[i][j] = 1;
                    } else {
                        // (i,j) is a square if (i-1,j-1) and (i-1, j) and (i, j-1) are all square
                        dp[i][j] = __min_3(dp[i-1][j-1], dp[i-1][j], dp[i][j-1]) + 1;
                    }
                    max_side = max(max_side, dp[i][j]);
                }
            }
        }
        return max_side*max_side;
    }
};
```
